```python
def make_new_columns(df,col_name, list_of_keys):
    """
    params will be(col_name,[list_of_keys]) ,
    the names in the list should be same as keys
    """
    if(col_name not in df.columns):
        print("No such column exists in the df")
    for key in list_of_keys:
        df[key] = ""
    for i in range(len(df)):
        response_list = eval(df["boxOffice"][i])
        if(response_list):
            response = response_list
            for it in response:
                for key in list_of_keys:
                    if key in it:
                        df[key][i] = it[key]["val"]
                        

    return df
    
    #delet the original column 
    
```

```python
new_df = df
```

```python
new_df = make_new_columns(new_df,"boxOffice", ["title-boxoffice-budget","title-boxoffice-cumulativeworldwidegross","title-boxoffice-grossdomestic","title-boxoffice-openingweekenddomestic"])
```

```python
def dollars_str_to_int(input_str):
    """
    Takes an input as a str. Saves currency signs if any
    saves value"""
    if(len(input_str)==0):
        return None
    #strip the str of any spaces
    stripped_input_str = input_str.strip()
    split_str = stripped_input_str.split()
    money_str = split_str[0]
    currency = money_str[:1]
    money_with_comma = money_str[1:]
    money = money_with_comma.replace(",","")
    return currency,money
```

```python
currency,money_int = dollars_str_to_int(new_df["title-boxoffice-budget"][2])
```
